```python
def message_schedule(block):
    # Chia block thành các chunk 32-bit
    chunk_size = 32  # Kích thước chunk là 32-bit
    num_chunks = 16  # Số lượng chunk trong message schedule

    # Chia chuỗi thành các chunk 32-bit
    chunks = [block[i:i+chunk_size] for i in range(0, len(block), chunk_size)]
    print("GIÁ TRỊ CỦA W_i ĐẦU TIÊN SAU KHI CHIA CÁC KHỐI:")
    for i in range(16):
        print("w[{}] = {}".format(i, chunks[i]))

    # Chuyển đổi từng chunk thành số nguyên 32-bit
    w = [0] * num_chunks
    for i in range(num_chunks):
        w[i] = int(chunks[i], 2)
    
    for i in range(3):
        for j in range(16):
            w.append(w[j])
    
    return w
```

```python
def f_MD5(x, y, z, i):
    if i >=0 and i <=15:
        return (x & y) | ((15+ ~x) & z)
    elif i >=16 and i <= 31:
        return (z & x) | ((15+ ~z) & y)
    elif i >= 32 and i <=48:
        return x ^ y ^ z
    else:
        return y ^ (x | (15 + ~z))

def ROTL(t, i):
    s = [7, 12, 17, 22] * 4 + [5, 9, 14, 20] * 4 + [4, 11, 16, 23] * 4 + [6, 10, 15, 21] * 4
    steps = s[i]
    t0 = str(bin(t))
    t0 = t0[2:]
    steps = steps % len(t0)
    t1 = t0[0: steps]
    t2 = t0[steps: ]
    return int(t2 + t1, 2)

def MD5(a, b, c, d, W, K):
    a_unit = a
    b_unit = b
    c_unit = c
    d_unit = d

    a1 = a
    b1 = b
    c1 = c
    d1 = d
    for i in range(64):
        a0 = a1
        b0 = b1
        c0 = c1
        d0 = d1

        c1 = b0
        d1 = c0
        a1 = d0
        t = (((a0 + f_MD5(b0, c0, d0, i)) % 2**32 + W[i]) % 2**32 + K[i]) % 2**32
        b1 = (b0 + ROTL(t, i)) % 2**32
    
        print("-----------------------------")
        print("LẦN LẶP THỨ {}:".format(i + 1))  
        print("a = ", dec_to_hex(a1))
        print("b = ", dec_to_hex(b1))
        print("c = ", dec_to_hex(c1))
        print("d = ", dec_to_hex(d1))
    
    a1 = a1 + a_unit
    b1 = b1 + b_unit
    c1 = c1 + c_unit
    d1 = d1 + d_unit
    return a1, b1, c1, d1

def dec_to_hex(a):
    a0 = str(hex(a))
    a0 = a0[2:]
    for i in range(8 - len(a0)):
        a0 = "0" + a0
    return a0
```

```python
def MD5_main(M_block_use, a, b, c, d):
    a1 = a
    b1 = b
    c1 = c
    d1 = d

    K = []
    for i in range(64):
        K.append(int(2**32 * abs(math.sin(i))))
    
    print("\n=========================================")
    print("GIÁ TRỊ CỦA K_i:")
    for i in range(64):
        print("K[{}] = {}".format(i, bin(K[i])))
    for i in range(len(M_block_use)):
        print("==========================================================")
        print("==========================================================")
        print("KHỐI THỨ {}:".format(i + 1))
        W = message_schedule(M_block_use[i])
        a0 = a1
        b0 = b1
        c0 = c1
        d0 = d1        
        a1, b1, c1, d1 = MD5(a0, b0, c0, d0, W, K)
        a1 = (a0 + a1) % 2**32
        b1 = (b0 + b1) % 2**32
        c1 = (c0 + c1) % 2**32
        d1 = (d0 + d1) % 2**32
        print("=========================================")
        print("GIÁ TRỊ H_i SAU KHI XỬ LÝ KHỐI THỨ {}:".format(i + 1))
        print("a = ", dec_to_hex(a1))
        print("b = ", dec_to_hex(b1))
        print("c = ", dec_to_hex(c1))
        print("d = ", dec_to_hex(d1))
        print("H[{}] = {}".format(i + 1, dec_to_hex(a1) + dec_to_hex(b1) + dec_to_hex(c1) + dec_to_hex(d1)))
    
    a2 = dec_to_hex(a1)
    b2 = dec_to_hex(b1)
    c2 = dec_to_hex(c1)
    d2 = dec_to_hex(d1)
    print("\n=========================================")
    print("GIÁ TRỊ BỘ ĐỆM H_i SAU KHI XỬ LÝ:")
    return a2 + b2 + c2 + d2
```
